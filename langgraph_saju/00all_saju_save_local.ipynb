# 모든 사주 문서들

```python
from langchain_community.document_loaders import TextLoader

file_paths = [
    "data/Simple-Chinese-Astrology_Combined.md",
    "data/The-Definitive-Book-of-Chinese-Astrology.md",
    "data/The-Four-Pillars-of-Destiny-Understanding-Character-Relationships-and-Potential-through-Chinese-Astrology.md",
    "data/all_lectures.md"
]

all_docs = []

for path in file_paths:
    loader = TextLoader(path, encoding="utf-8")
    docs = loader.load()
    all_docs.extend(docs)  # 각 파일에서 로드한 docs를 리스트에 추가

print(len(all_docs))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

documents = text_splitter.split_documents(docs)
len(documents)
```

```python
from langchain.vectorstores import FAISS
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="bge-m3")

vector_store = FAISS.from_documents(embedding=embeddings, documents=documents)
vector_store.save_local("faiss_saju/all_saju_data")
```

```python
retriever = vector_store.as_retriever(kwargs={"k":10})
```

```python
db = vector_store.load_local("faiss_saju/all_saju_data", embeddings=embeddings, allow_dangerous_deserialization=True)
```

```python
retriever = db.as_retriever(kwargs={"k":10})
```

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import CrossEncoderReranker
from langchain_community.cross_encoders import HuggingFaceCrossEncoder

model = HuggingFaceCrossEncoder(model_name="BAAI/bge-reranker-base")
compressor = CrossEncoderReranker(model=model, top_n=10)
compression_retriever = ContextualCompressionRetriever(
    base_compressor=compressor, base_retriever=retriever
)
```

```python
#from langchain import hub
from langchain_core.prompts import load_prompt

from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

from operator import itemgetter

# prompt = hub.pull("teddynote/rag-prompt-chat-history")
prompt = load_prompt("../../prompt/saju-rag-promt_korea.yaml")

# model = ChatOpenAI(model="gpt-4o")
from langchain_google_genai import ChatGoogleGenerativeAI

#            "gemini-1.5-pro" in self.model
#            or "gemini-1.5-flash" in self.model
#            or "gemini-2" in self.model

model = ChatGoogleGenerativeAI(model="gemini-2.0-flash",temperature=0.1)

pdf_chain = {
    "question": itemgetter("question"),
    "context": itemgetter("context"),
    "chat_history": itemgetter("chat_history"),
} | prompt | model | StrOutputParser()
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory

# 세션 기록을 저장할 딕셔너리
store = {}

# 세션 ID를 기반으로 세션 기록을 가져오는 함수
def get_session_history(session_ids):
    print(f"[대화 세션ID]: {session_ids}")
    if session_ids not in store:  # 세션 ID가 store에 없는 경우
        # 새로운 ChatMessageHistory 객체를 생성하여 store에 저장
        store[session_ids] = ChatMessageHistory()
    return store[session_ids]  # 해당 세션 ID에 대한 세션 기록 반환

# 대화를 기록하는 RAG 체인 생성
rag_with_history = RunnableWithMessageHistory(
    pdf_chain,
    get_session_history,  # 세션 기록을 가져오는 함수
    input_messages_key="question",  # 사용자의 질문이 템플릿 변수에 들어갈 key
    history_messages_key="chat_history",  # 기록 메시지의 키
)
```

```python
question = "내 생년월일은 양력 1995년 3월 28일 남자 12지시중 오시에 태어났어. 전반적인 사주봐줘"
response = rag_with_history.invoke({"question": question, "context":compression_retriever.invoke(question)}, config={"configurable": {"session_id": "rag123"}},)
print(response)
```
